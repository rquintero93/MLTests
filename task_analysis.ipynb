```python
sales=pd.read_csv('sales.csv',header=0)
sales
```

```python
merged=pd.merge(tasks,sales,how='right',on='Agent')
merged
```

```python
merged.fillna(0,inplace=True)
```

```python
merged.to_csv('merged.csv',index=False)
```

```python

merged_numeric=merged.select_dtypes(include=[np.number])
mask=np.triu(merged_numeric.corr())
plt.figure(figsize=(12,12))
sns.heatmap(merged_numeric.corr(),annot=True,cmap='coolwarm',fmt='.2f',mask=mask,vmax=1,vmin=-1,square=True)
plt.show()
```

```python
merged_pairplot=merged_numeric.drop(['MedianConnectedCallTime','GrossCRLead_per','L2A_per','A2PA_per','AverageTicket'],axis=1)
sns.pairplot(merged_pairplot,corner=True)
```

```python
task_schools=pd.read_csv('tasks_w_schools.csv',header=0)
task_schools
```

```python
sales_schools=pd.read_csv('sales_w_schools.csv',header=0)
sales_schools
```

```python
merged_schools=pd.merge(task_schools,sales_schools,how='right',on=['Agent','SalesCoordinator','ic_school_name'])
merged_schools
```

```python
merged_schools.fillna(0,inplace=True)
merged_schools.rename(columns={'ic_school_name':'School'},inplace=True)
merged_schools
```

```python
values_to_drop = [0, "GA - School 1 Test", "GA - School 2 Test", "GA School 1 Test"]
merged_schools = merged_schools[~merged_schools['School'].isin(values_to_drop)]
merged_schools
```

```python
schools_agg=merged_schools.groupby('School').agg({'TotalCalls':'sum','UsefulCalls':'sum','ConnectedCalls':'sum','TotalEmails':'sum','ConnectedCallTimeMins':'sum','UsefulCallTimeMins':'sum','Booking':'sum','TotalRevenue':'sum','AllApps':'sum','PaidApps':'sum','TotalNewLeads':'sum'})
schools_agg.reset_index(inplace=True)
schools_agg
```

```python
# Drop rows with specific values
values_to_drop = [0, "GA - School 1 Test", "GA - School 2 Test", "GA School 1 Test"]
schools_agg = schools_agg[~schools_agg['School'].isin(values_to_drop)]
schools_agg
```

```python
ax=sns.scatterplot(data=merged_schools,x='TotalEmails',y='TotalRevenue',hue='School',palette='Set2')

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
ax=sns.scatterplot(data=merged_schools,x='ConnectedCalls',y='TotalRevenue',hue='School',palette='Set2')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
ax=sns.jointplot(data=merged_schools,x='TotalEmails',y='TotalRevenue',kind='reg')

ax.figure.set_size_inches(12,12)

plt.show()
```

```python
ax=sns.jointplot(data=merged_schools,x='ConnectedCalls',y='TotalRevenue',kind='reg')

ax.figure.set_size_inches(12,12)

plt.show()
```
